```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Data Preprocessing

```
dataset = pd.read_csv('Market_Basket_Optimisation.csv', header = None)
```

```
transactions = []
for i in range(0, dataset.shape[0]):
    transaction = [str(dataset.values[i, j]) for j in range(0, dataset.shape[1]) if str(dataset.values[i, j]) != 'nan']
    transactions.append(transaction)
```

## Training the Apriori model on the dataset

```
from apyori import apriori
rules = apriori(transactions = transactions, min_support = 0.003, min_confidence = 0.2, min_lift = 3,min_length = 2,max_length = 2)
```

## Visualising the results

### Displaying the first results coming directly from the output of the apriori function

```
res = list(rules)
res
```

### Putting the results well organised into a Pandas DataFrame

```
def inspect(results):
    lhs         = [tuple(result[2][0][0])[0] for result in results]
    rhs         = [tuple(result[2][0][1])[0] for result in results]
    supports    = [result[1] for result in results]
    confidences = [result[2][0][2] for result in results]
    lifts       = [result[2][0][3] for result in results]
    return list(zip(lhs, rhs, supports, confidences, lifts))
resultsinDataFrame = pd.DataFrame(inspect(res), columns = ['Left Hand Side', 'Right Hand Side', 'Support', 'Confidence', 'Lift'])
```

### Displaying the results non sorted

```
resultsinDataFrame
```

### Displaying the results sorted by descending lifts

```
resultsinDataFrame.nlargest(n = 10, columns='Lift')
```

```
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.scatterplot(
    x='Support',
    y='Confidence',
    hue='Lift',
    size='Lift',
    sizes=(50, 500), # Adjust point size range as needed
    data=resultsinDataFrame,
    palette='viridis', # Choose a color palette
    legend='full'
)

plt.title('Association Rules: Support vs. Confidence (Colored by Lift)')
plt.xlabel('Support')
plt.ylabel('Confidence')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```
